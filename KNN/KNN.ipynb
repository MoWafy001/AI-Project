```python
dataset_path = '../Datasets/SVM-main-data/Social_Network_Ads.csv'
```

**Load Data**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv(dataset_path)
df.head()
```

**Print unique Target values**

```python
# Extract the values of the 'Purchased' column
target_values = df['Purchased'].values

# Print the target values
np.unique(target_values)
```

**Data Type of Columns**

```python
df.dtypes
```

```python
df.head()
```

```python
df['Gender_into_nums'] = pd.factorize(df.Gender)[0]
```

```python
titles = list(df.columns)
titles
```

**Swapping columns**

```python
titles[4], titles[5] = titles[5], titles[4]
titles[3], titles[4] = titles[4], titles[3]
titles[2], titles[3] = titles[3], titles[2]
```

```python
titles
```

```python
df = df[titles]
df
```

```python
df.shape
```

```python
X = df.iloc[:, [2,3,4]]
Y = df.iloc[:, 5]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Splitting the dataset into the Training and Test set
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
print("Training data : ",X_Train.shape)
print("Testing data : ",X_Test.shape)
```

**Learning (Training)**

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=11)
KNN.fit(X_Train, Y_Train)
```

```python
Y_Pred = KNN.predict(X_Test)
```

```python
Y_Pred
```

```python
len(Y_Pred)
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(Y_Test, Y_Pred)
print(accuracy)
```
